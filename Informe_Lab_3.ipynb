##**Gráficas de Ejecución**

A continuación se muestran gráficas para la ejecución del programa con 1, 2, 4 y 8 hilos, en relación a las ejecuciones realizadas y su rendimiento en ms.

```python
import matplotlib.pyplot as plt

hilo = [2940, 3192, 3201, 3175, 3147, 2982, 3036, 3164, 3157, 3177]

fig, ax = plt.subplots()

ax.plot(hilo, label=f"Tamaño vector = 100000, 1000 iteraciones")

plt.xlabel('Ejecuciones')
plt.ylabel('Milisegundos')
plt.title(label="Rendimiento de 1 hilo en ms")
plt.legend()
plt.show()
```

#**Comparación entre hilos**

Para tener una mejor visualización de como se compara el rendimiento de los hilos en su ejecución con diferentes iteraciones:

```python

num_hilos = [1, 2, 4, 8]
ejecuciones = list(range(1, 11))  # Número de iteraciones

plt.figure(figsize=(10, 6))

plt.plot(ejecuciones, hilo, marker='o', label='1 hilo')
plt.plot(ejecuciones, dos_hilos, marker='o', label='2 hilos')
plt.plot(ejecuciones, cuatro_hilos, marker='o', label='4 hilos')
plt.plot(ejecuciones, ocho_hilos, marker='o', label='8 hilos')

plt.xlabel('Ejecuciones')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Comparacion entre hilos')
plt.legend()

plt.show()
```

Podemos ver que para el hilo 1 la ejecución en tiempo es estable durante todas las iteraciones, para 2 hilos la ejecución se mantiene estable en la mayoría de ejecuciones excepto en tres donde el tiempo aumenta considerablemente, algo similar ocurre para el manejo de 4 y 8 hilos. Sin embargo, también es notable que el rendimiento con 8 hilos es mayor al de resto de hilos en la mayoría de ejecuciones.

##**Conclusiones**



*   Al analizar la gráfica es posible ver que el rendimiento en ejecución en cuanto a la cantidad de hilos, ocupa un menor tiempo o es más óptimo cuando la cantidad de hilos aumenta, esto puede deberse a como se usan los recursos del equipo con cada uno de los hilos, al hacer uso de los recursos en la ejecución de un solo hilo, puede haber un desgaste de los mismos, donde el sistema no puede aprovechar plenamente la capacidad de procesamiento disponible.
